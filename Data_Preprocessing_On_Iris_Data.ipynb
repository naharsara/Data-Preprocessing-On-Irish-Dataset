# Handling Duplicate values

```python
# Check for duplicates
duplicates = iris[iris.duplicated()]
print("Number of duplicate rows:", len(duplicates))

# Drop duplicates
df = iris.drop_duplicates()

# Verify that duplicates have been removed
duplicates = df[df.duplicated()]
print("Number of duplicate rows after removal:", len(duplicates))
```

```python
df
```

# Columns Cleanup


```python


# Rename columns
df = df.rename(columns={
    'sepal_length': 'input1',
    'sepal_width': 'input2',
    'petal_length': 'input3',
    'petal_width': 'input4',
    'species': 'Output'
})

# Verify column names
print(df.head())
```

# DataFrame slicing


```python
# Slice based on rows using loc[]
x = df.loc[30:40, ['input1']]

# Slice based on rows using iloc[] index
y = df.iloc[30:40, 2:3]

print("Slice using loc[]:")
print(x)

print("Slice using iloc[]:")
print(y)
# Slice based on conditions using loc[]
z = df.loc[df['input3'] > 2.0, ['input1', 'Output']]

print("Slice using loc[] and condition:")
print(z)
```

# selecting, extracting

```python
# Select a single column
x = df['input1']

print("Selected column 'input1':")
print(x)
# Select multiple columns
y = df[['input1', 'input2']]

print("Selected columns 'input1' and 'input2':")
print(y)
# Extract rows based on conditions
setosa = df.loc[df['Output'] == 'Iris-setosa']

print("Rows with Output 'Iris-setosa':")
print(setosa.head(20))
```
